# API  ( Application Programing Interface)
## 1) An API is a server that you can use to retrieve and send data .API are most commanly used to retrieve data.
## 2) An API acts as acommunication layer , or as the name says , an interface , that allows differnt systems to talk to each other
## 3) We have to make a request for information or data , and the API returns a response with what is the requested . When we want to recive data from the API , we need to make a request . Sending a request to an API is known as calling an API.
## )4 Response returned from API is generally in JSON ( Java Script Object Notation ) format

#  JSON Place Holder Posts
## URL stands for Uniform Resource Locator

```python
url1 = "https://jsonplaceholder.typicode.com/posts/"
res = requests.get(url1)
print(res)
```

```python
print(res.json())
```

```python
r1 = res.json()
print(type(r1))
print(len(r1))
```

```python
d = [{'a1':10,'a2':20},{'a1':30,'a2':40},{'a1':50,'a2':70}]
for i in d:
    print(i['a1'],i['a2'])
```

```python
for i in r1:
    print('UserID:',i['userId'],'Id:',i['id'],'\nTitle:',i['title'],'\nBody:',i['body'])
    print('-'*70)
```

```python
userId,Id = [],[]
title,body = [],[]
for i in r1:
    userId.append(i['userId'])
    Id.append(i['id'])
    title.append(i['title'])
    body.append(i['body'])
```

```python
print(len(userId),len(Id),len(title),len(body))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'UserId':userId,'ID':Id,'Title':title,'Body':body})
df.head()
```
