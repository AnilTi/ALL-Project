# Violin Plot using Seaborn in Python :-

A violin plot plays a similar activity that is pursued through whisker or box plot do. As it shows several quantitative data across one or more categorical variables. It can be an effective and attractive way to show multiple data at several units.

## Example 1: Basic visualization of “fmri” dataset using violinplot()  :-

Functional magnetic resonance imaging or functional MRI (fMRI) measures brain activity using a strong, static magnetic field to detect changes associated with blood flow. When an area of the brain is in use, blood flow to that region also increases. The increased blood flow is represented by a higher-amplitude signal seen as strong neural activity.

```python
import seaborn
import matplotlib.pyplot as plt
```

```python
seaborn.set(style = 'whitegrid')
fmri = seaborn.load_dataset("fmri")
```

```python
fmri.head()
```

```python
seaborn.violinplot(x ="timepoint",y ="signal",data = fmri)
plt.show()
```

## Example 2: Grouping data points on the basis of category, here as region and event :-

```python
seaborn.violinplot(x ="timepoint",y ="signal",hue ="region",style ="event",data = fmri)
plt.show()
```

## Example 3: Basic visualization of “tips” dataset using violine plot():-

```python
seaborn.set(style = 'whitegrid') 
tip = seaborn.load_dataset('tips')
seaborn.violinplot(x ='day', y ='tip', data = tip)
plt.show()
```

```python
seaborn.set(style = 'whitegrid') 
tip = seaborn.load_dataset('tips')
seaborn.boxplot(x ='day', y ='tip', data = tip)
plt.show()
```

## 1.Draw a single horizontal plot using only one axis:

```python
# use to set style of background of plot
seaborn.set(style="whitegrid")
 
# loading data-set
tips = seaborn.load_dataset("tips")
 
seaborn.violinplot(x=tip["total_bill"])
plt.show()
```

## 2. Draw horizontal violinplot:-

```python
# use to set style of background of plot
seaborn.set(style="whitegrid")
 
# loading data-set
tips = seaborn.load_dataset("tips")
 
seaborn.violinplot(x="tip", y="day", data=tip)
plt.show()
```

## 3. Using hue parameter:-

```python
# use to set style of background of plot
seaborn.set(style="whitegrid")
 
# loading data-set
tips = seaborn.load_dataset("tips")
 
seaborn.violinplot(x="day", y="total_bill", hue="time", data=tips)
plt.show()
```

## 4. Draw outlines around the data points using linewidth:

```python
# use to set style of background of plot
seaborn.set(style="whitegrid")
 
# loading data-set
tips = seaborn.load_dataset("tips")
 
seaborn.violinplot(x ='day', y ='tip', data = tip, linewidth = 4)
plt.show()
```

### 5. Draw each level of the hue variable at different locations on the major categorical axis:

When using hue nesting, setting dodge should be True will separate the point for different hue levels along the categorical axis. And Palette is used for the different levels of the hue variable.

```python
# use to set style of background of plot
seaborn.set(style="whitegrid")
 
# loading data-set
tips = seaborn.load_dataset("tips")
 
seaborn.violinplot(x="day", y="total_bill", hue="smoker",
                   data=tips, palette="Set2", dodge=True)

plt.show()
```

## 7. Control violin order by passing an explicit order:

```python
# use to set style of background of plot
seaborn.set(style="whitegrid")
 
# loading data-set
tips = seaborn.load_dataset("tips")
 
seaborn.violinplot(x="time", y="tip", data=tips,
                   order=["Dinner", "Lunch"])

plt.show()
```

## 8. Adding color parameter:

```python
# use to set style of background of plot
seaborn.set(style="whitegrid")
 
# loading data-set
tips = seaborn.load_dataset("tips")
 
seaborn.violinplot(x ='day', y ='tip', data = tip, color = "Yellow")
plt.show()
```

### 9. Scale the violin width by the number of observations in each bin:

```python
# use to set style of background of plot
seaborn.set(style="whitegrid")
 
# loading data-set
tips = seaborn.load_dataset("tips")
 
seaborn.violinplot(x="day", y="total_bill", hue="sex",
                    data=tip, palette="Set2", split=True,
                    scale="count")
plt.show()
```
