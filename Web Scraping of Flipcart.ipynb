```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url='https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
```

```python
req=requests.get(url)
```

```python
req
```

```python
# Show data in html form but not structured manner for
#  Solving this problem we have used BeautifulSoup library.
```

```python
content=BeautifulSoup(req.content,'html.parser')
#content
print(content.prettify())
```

```python
laptops=content.find_all('div',{'class':'_4rR01T'})
```

```python
print(laptops)
```

```python
print(len(laptops))
```

```python
laptop=laptops[0] # 1st laptop information
laptop
```

```python
# Name of Laptop
name=content.find_all('div',{'class':'_4rR01T'})
#print(name)
#print(type(name))
#name=name[0]
#print(name)
#name=name.text
#print(name)
```

```python
# Rating of Laptop
rating=content.find_all('div',{'class':'_3LWZlK'})
#rating=rating[0].text
#print(rating)
```

```python
# Price of Laptop
price=content.find_all('div',{'class':'_30jeq3 _1_WHN1'})
#print(price)
#price=price[0].text
#print(price)
```

```python
nm=[]
pr=[]
rt=[]
```

```python
for i in name:
    nm.append(i.text) 
    
for i in price:
    pr.append(i.text) 
    
for i in range(len(nm)):
    rt.append(rating[i].text)
```

```python
data={'NAME':nm,'PRICE':pr,'RATING':rt}
df=pd.DataFrame(data)
```

```python
print(df)
```

```python
df.to_csv("laptops.csv",encoding="utf-8", index=False)
```
