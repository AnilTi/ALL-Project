```python
# NOW WEB SCRAPING FROM REGULAR EXPRESSION
```

```python
import re
```

```python
res=requests.get('https://www.tgcindia.com/blog/')
```

```python
print(res.text)
```

```python
# How to find out all links
```

```python
def find_links(page_source):
    regex=re.compile(r'<a href="(.+?)"')
    result=regex.findall(page_source)
    return result
```

```python
links=find_links(res.text)
print(links)
```

```python
li=[]
for link in links:
    if link!='#' and link!='#login':
        li.append(link)
        print(link)
```

```python
# How to find out Top 10 links
```

```python
li
```

```python
top_links=li[0:5]
top_links
```

```python
# How to open top 10 links in Web browser automaticaly
```

```python
import webbrowser
```
